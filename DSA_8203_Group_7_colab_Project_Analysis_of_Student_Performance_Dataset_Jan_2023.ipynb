```python
passing_marks = 40
Student['Pass_Status'] = np.where(Student['total_score']<passing_marks,'F','P')
Student['Pass_Status'].value_counts()
```

```python
Student.head(2)
```

```python
#Assigning Grades.
# Function to assign grades.

def get_grade(marks):
    if marks >= 82 and marks <= 100:
        return 'A+'
    elif marks >=73 and marks < 82:
        return 'A'
    elif marks >=64 and marks < 73:
        return 'B+'
    elif marks >= 55 and marks < 64:
        return 'B'
    elif marks >=46 and marks < 55:
        return 'C'
    elif marks >= 35 and marks < 46:
        return 'P'
    elif marks < 35:
        return 'F'
```

```python
#Assigning Grades Continuation.
Student['Reading Grade'] = Student['reading score'].apply(get_grade)
Student['Writing Grade'] = Student['writing score'].apply(get_grade)
Student['Math Grade'] = Student['math score'].apply(get_grade)
```

```python
def grades(pass_status,total_score):
    if pass_status == 'F':
        return 'F'
    if total_score >=90:
        return 'A+'
    if total_score >=80:
        return 'A'
    if total_score >=70:
        return 'B'
    if total_score >=60:
        return 'C'
    if total_score >=50:
        return 'D'
    if total_score >=40:
        return 'E'
```

```python
Student['Grade'] =Student.apply(lambda x : grades(x['Pass_Status'] ,x['total_score']) ,axis =1)
Student.head(2)
```

```python
#Our next step is to divide the data into attributes and labels.
#Attributes are the independent variables while labels are dependent variables whose values are to be predicted.
#We want to predict the total_score depending upon the math score recorded. 
X_shaped=Student["math score"].values.reshape(-1,1)
Y_shaped=Student["total_score"].values.reshape(-1,1)
```

```python
#random state can be any number.
#Next, we split 80% of the data to the training set while 20% of the data to test set using below code.
#The test_size variable is where we actually specify the proportion of the test set.
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X_shaped,Y_shaped,test_size=0.2,random_state=0)
print(x_train.shape)
print(x_test.shape)
print(y_train)
print(y_test)
```

Initializing the model

```python
from sklearn.linear_model import LinearRegression
lm=LinearRegression()
```

Fitting the model

```python
#After splitting the data into training and testing sets, finally, the time is to train our algorithm.
lm.fit(x_train,y_train)
```

```python
LinearRegression()
```

```python
#finds the best value for the intercept.
Intercept=lm.intercept_
print(f"the y-intercept is:{Intercept}")
```

```python
#finds the best value for the Coefficient.
Coefficient=lm.coef_
print(f"the Coefficient is:{Coefficient}")
```

This means that for every one unit of change in math score, the change in the total_score is about 0.85%.

Now that we have trained our algorithm, itâs time to make some predictions. To do so, we will use our test data and see how accurately our algorithm predicts the percentage score.

```python
# making the predictions using the predict function.
y_pred = lm.predict(x_test)
```

The straight line in the above graph shows our algorithm is correct.

The final step is to evaluate the performance of the algorithm. This step is particularly important to compare how well different algorithms perform on a particular dataset. For regression algorithms, three evaluation metrics are commonly used

```python
from sklearn import metrics
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))  
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))  
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

You can see that the value of root mean squared error is 6.01, which is more than 10% of the mean value of the percentages of all the scores. This means that our algorithm was not very accurate but can still make reasonably good predictions.
