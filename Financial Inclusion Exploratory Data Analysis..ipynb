```python
Fp["age_of_respondent"] = Fp["age_of_respondent"].apply(lambda x: float(x))
```

```python
gender = Fp.loc[:,['gender_of_respondent','age_of_respondent']]
```

```python
gender
```

```python
gender['mean_age'] = gender.groupby('gender_of_respondent')['age_of_respondent'].transform("mean")
```

```python
gender
```

```python
gender.drop('age_of_respondent', axis=1, inplace=True)
```

```python
gender
```

```python
gender = gender.drop_duplicates().sort_values('mean_age', ascending=False)
gender
```

```python
#Mean Age pie graph.
#50.3% males have a mean age of 39 years.49.7% females have a mean of 38 years.
##we have more males than females implying that majority of the males in the dataset are older than females.
sizes=gender['mean_age']
fig = plt.figure(figsize=(10,7))
plt.title("Mean Age Pie Chart")
explode = (0, 0.1) 
plt.pie(sizes,explode,labels=gender['gender_of_respondent'], autopct='%1.1f%%', shadow=True)

plt.axis('equal')
plt.legend()
plt.show()
```

```python
#Mean Age pie graph.
#50.3% males have a mean age of 39 years.49.7% females have a mean of 38 years.
#we have more males than females implying that males in the dataset are older than females.
import plotly.express as px
fig = px.pie(gender, names='gender_of_respondent', values='mean_age', template='seaborn')
fig.update_traces( textinfo="percent+label")
fig.show()
```
